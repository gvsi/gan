```python
from json_tricks import dumps, loads
from tqdm import tqdm
```

```python
i = 0
lines = []
with open("data/data4_valid_mul5/fl4-mul5-comb-new.txt") as f:
    for line in f:
        lines.append(line)
```

```python
scores = {}
for i in tqdm(range(len(lines))):
    map_str_comb, dic = lines[i].split("\t")
    obj = loads(dic)
    scores[map_str_comb] = obj['valid_score']
```

```python
len(scores.items())
```

```python
i = 0
lines2 = []
with open("data/data4_valid_mul5_gan/fl4-mul5-gan-comb-new.txt") as f:
    for line in f:
        lines2.append(line)
```

```python
scores2 = {}
for i in tqdm(range(len(lines2))):
    map_str_comb, dic = lines2[i].split("\t")
    obj = loads(dic)
    scores2[map_str_comb] = obj['valid_score']
```

```python
print(len(scores2))

scores2
```

```python
diffs = 0
percs = 0
tot = 0
tot_perc = 0
for map_str, score in scores2.items():
    if score > 0:
        diffs += score - scores[map_str]
#         print(str(map_str) + "\t" + str(scores[map_str]) + "\t" + str(score) + "\t" + str(score >= scores[map_str]))
        tot += 1
    if score > 0 and scores[map_str] > 0:
        percs += (score - scores[map_str]) / scores[map_str]
        tot_perc += 1
    
print(diffs / tot)
print(tot)
print(percs / tot_perc)
print(tot_perc)
```
